```python
import pandas as pd
import gdown


# ID for Chicago Dataset
chicago_file_id = '1CaIk9eAs9fDKhTGIUqCHImypDidDbj_1'

# ID for New Orleans Dataset
new_orleans_file_id = '1Ebmj5Wiw365oYEr0C3prIYN_QVlSHt-V'


print(" Downloading Chicago Data...")
url_chicago = f'https://drive.google.com/uc?id={chicago_file_id}'
gdown.download(url_chicago, 'chicago_data.csv', quiet=False)

print("\n⬇ Downloading New Orleans Data...")
url_no = f'https://drive.google.com/uc?id={new_orleans_file_id}'
gdown.download(url_no, 'new_orleans_data.csv', quiet=False)

# --- STEP 3: Read and Label them ---

# Read the CSVs
chicago_df = pd.read_csv('chicago_data.csv')
new_orleans_df = pd.read_csv('new_orleans_data.csv')

# Add a 'City' column so we know which row belongs to which city
chicago_df['City'] = 'Chicago'
new_orleans_df['City'] = 'New Orleans'

# --- STEP 4: Merge into one Master File ---

# Concatenate (stack them on top of each other)
df = pd.concat([chicago_df, new_orleans_df], ignore_index=True)

# Show the result
print(f"\nSuccess! Merged dataset has {len(df)} rows.")
print("Here is a preview:")
df.head()
```

```python
# 1. Clean the symbols again (just to be safe)
# We treat the column as a string, replace '$' and ',' with nothing.
df['price'] = df['price'].astype(str).str.replace('$', '').str.replace(',', '')

# 2. Convert to numeric with 'coerce'
# This is the MAGIC FIX. It turns the string "nan" back into a proper null value.
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# 3. Check how many prices are missing
missing_count = df['price'].isnull().sum()
print(f"Prices cleaned! Rows with missing prices: {missing_count}")

# 4. (Recommended) Drop rows where Price is missing
# A listing with no price is useless for our analysis, so let's remove them.
df = df.dropna(subset=['price'])

print("Price column fixed and ready.")
df.head()
```

```python
# 1. Check the data type first
print("Before cleaning:", df['price'].dtype)

# 2. Remove '$' and ',' signs
# We treat the column as a string (.str), replace the symbols with nothing, and convert to float.
df['price'] = df['price'].astype(str).str.replace('$', '').str.replace(',', '')
df['price'] = pd.to_numeric(df['price'])

# Check again
print("After cleaning:", df['price'].dtype)
```

```python
# 1. Fill missing review data with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['number_of_reviews'] = df['number_of_reviews'].fillna(0)

# 2. Drop rows that are missing essential identity info
# We are dropping rows where 'host_name' or 'name' is missing
df.dropna(subset=['host_name', 'name'], inplace=True)

# 3. Check if any nulls remain (Should be 0 for these columns)
print("Remaining Nulls in Reviews:", df['reviews_per_month'].isnull().sum())
print("Remaining Nulls in Host Name:", df['host_name'].isnull().sum())
```

```python
# --- Transformation 1: Price Category ---
def categorize_price(price):
    if price < 100:
        return 'Budget (<$100)'
    elif price <= 300:
        return 'Standard ($100-$300)'
    else:
        return 'Premium (>$300)'

df['Price Category'] = df['price'].apply(categorize_price)

# --- Transformation 2: Host Type ---
# Count how many listings each host has
host_counts = df['host_id'].value_counts()

# Map that count back to the original dataframe
df['Host Listings Count'] = df['host_id'].map(host_counts)

# Create the label
def categorize_host(count):
    if count == 1:
        return 'Single Host'
    else:
        return 'Multi-Listing Host'

df['Host Type'] = df['Host Listings Count'].apply(categorize_host)

# Preview the new magic columns
print(df[['name', 'price', 'Price Category', 'Host Type']].head())
```

```python
# --- THE GEOGRAPHIC FIX ---
# 1. Convert lat/long to numeric (just in case they are strings)
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

# 2. Drop rows where Latitude or Longitude is missing
# This removes the "Null Island" (Africa) problem!
df = df.dropna(subset=['latitude', 'longitude'])

# 3. Final Check
print(f"Latitude Nulls: {df['latitude'].isnull().sum()}")
print(f"Longitude Nulls: {df['longitude'].isnull().sum()}")

# Save and download the fixed file
df.to_csv('Final_Airbnb_Capstone_Data.csv', index=False)
from google.colab import files
files.download('Final_Airbnb_Capstone_Data.csv')
```
